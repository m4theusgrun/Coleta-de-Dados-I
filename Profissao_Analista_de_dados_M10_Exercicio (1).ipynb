## 1\. Excel para CSV

Utilizando o pacote Python `openpyxl` visto em aula, extraia os seguintes as colunas `id`, `sexo` e `idade` para dos clientes inadimplentes (`default = 1`) e solteiros (`estado_civil = 'solteiro'`). Salves os dados extraídos no arquivo csv `credito.csv` separado por `;`. Exemplo do cabeçalho e das três primeiras linhas:

```
id;sexo;idade
767712558;59;M
713741358;46;M
772390908;59;M
```

**Dica:** O arquivo csv `credito.csv` deve ter 669 linhas, contando com o cabeçalho.

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````
# solução do exercício 1
from openpyxl import load_workbook

planilhas = load_workbook(filename='credito.xlsx')
planilha = planilhas.active

cabecalho = next(planilha.values)
extrair_id = cabecalho.index('id')
extrair_sexo = cabecalho.index('sexo')
extrair_idade = cabecalho.index('idade')
inadimplente = cabecalho.index('default')
solteiro = cabecalho.index('estado_civil')

dados_extraidos = []

for linha in planilha.values:

    if linha[inadimplente] == 1 and linha[solteiro] == 'solteiro':
        dados_extraidos.append([linha[extrair_id], linha[extrair_sexo], linha[extrair_idade]])

print(dados_extraidos)
````

---

````
import csv

with open(file='credito.csv', mode='w', encoding='utf8') as credito:
  escrita = csv.writer(credito, delimiter= ';')
  escrita.writerow(['id', 'sexo', 'idade'])
  escrita.writerows(map(lambda linha: linha, dados_extraidos))
````

**Dica:** Sua solução deve gerar o dicionário Python `credito` igual ao exemplo mas a ordem dos elementos pode variar tranquilamente.

**Dica:** Uma excelente forma de remover elementos duplicados de uma lista é convertê-la para `set` e depois para `list` novamente.



````
# solução do exercício 2
!pip install openpyxl
from openpyxl import load_workbook

workbook = load_workbook('credito.xlsx')
sheet = workbook.active

tipo_cartao_set = set()
escolaridade_set = set()

for row in sheet.iter_rows(min_row=2, values_only=True):
    tipo_cartao, escolaridade = row[0], row[1]

    tipo_cartao_set.add(tipo_cartao)
    escolaridade_set.add(escolaridade)

tipo_cartao_list = sorted(list(tipo_cartao_set))
escolaridade_list = sorted(list(escolaridade_set))

credito = {
    'tipo_cartao': tipo_cartao_list,
    'escolaridade': escolaridade_list
}

print(credito)
````

## 3\. Texto para CSV

No arquivo de texto `ebac.txt` você encontra o texto presente no rodapé da página de cursos da EBAC ([link](https://ebaconline.com.br/)).



**Arquivo TXT:** ebac.txt

````
%%writefile ebac.txt
MÍDIAS SOCIAIS
Instagram, Facebook, Youtube, LinkedIn

CURSOS
Software, Design, Marketing, Audiovisual, Programação & Data, Games

WEBINARS
Próximos, Anteriores

SOBRE
Sobre nós, Centro de carreiras, Vagas

CONTATO
WhatsApp +55 (11) 4200-2991
Telefone +55 (11) 3030-3200

BLOG
Design, Audiovisual, Marketing
````

Extraia os números de contato do arquivo texto `ebac.txt` e salve-os no arquivo csv `contato_ebac.csv` com o separador `;` no seguinte formato:

```
tipo;numero
whatsapp;+551142002991
telefone;+551130303200
```

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````
# solução do exercício 3 (bônus)
import csv
import re

def extrair_numeros_de_contato(texto):
    contatos = []

    padrao_whatsapp = re.compile(r'\+55\d{11}')
    padrao_telefone = re.compile(r'\+55\d{10}')

    whatsapp_numeros = padrao_whatsapp.findall(texto)
    for numero in whatsapp_numeros:
        contatos.append(('whatsapp', numero))

    telefone_numeros = padrao_telefone.findall(texto)
    for numero in telefone_numeros:
        contatos.append(('telefone', numero))

    return contatos

with open(file= 'ebac.txt', mode= 'r', encoding= 'utf8') as arquivo:
    linhas = arquivo.read()

    contatos = extrair_numeros_de_contato(linhas)

with open(file= 'contato_ebac.csv', mode= 'w', encoding= 'utf8', newline= '') as arquivo_csv:
    escrita = csv.writer(arquivo_csv, delimiter= ';')
    escritor_csv = csv.writer(arquivo_csv, delimiter=';')
    escritor_csv.writerow(['tipo', 'numero'])
    escritor_csv.writerows(contatos)

print('Dados salvos em contato_ebac.csv')
````
